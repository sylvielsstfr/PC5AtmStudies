# Input

```python
df_snls=pd.read_csv('SNLS_catD1D2D4.csv',index_col=0)
```

```python
#df_sed=pd.read_csv("regenerated_pickle_uvi_snlsmagcol.csv",index_col=0)
df_sed=pd.read_csv("regenerated_pickle_uvk_snlsmagcol.csv",index_col=0)
```

```python
df_snls["mxu"].hist(bins=100,color='blue',alpha=0.8,normed=1,label="SNLS")
df_sed["u"].hist(bins=100,color='red',alpha=0.8,normed=1,label="Pickles")

plt.xlabel("U (mag)",fontsize=20,fontweight='bold')
plt.title("SNLS/Regenerated Pickles UVI : U magnitudes",fontsize=25,fontweight='bold')
plt.legend()
```
