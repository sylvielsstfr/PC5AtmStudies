# Read file

```python
df=pd.read_csv('SNLS_catD1D2D4.csv',index_col=0)
```

```python
df1=df[df.data=='D1']
df2=df[df.data=='D2']
df3=df[df.data=='D4']
```

```python
new_ra1 = coord.Angle(df1["rag"], unit=u.degree) # create an Angle object
new_ra1.degree # convert to degrees
new_ra1 = new_ra1.wrap_at(180*u.degree)
new_dec1 = coord.Angle(df1["decg"], unit=u.degree)
```

```python
new_ra2 = coord.Angle(df2["rag"], unit=u.degree) # create an Angle object
new_ra2.degree # convert to degrees
new_ra2 = new_ra2.wrap_at(180*u.degree)
new_dec2 = coord.Angle(df2["decg"],unit=u.degree)
```

```python
new_ra3 = coord.Angle(df3["rag"], unit=u.degree) # create an Angle object
new_ra3.degree # convert to degrees
new_ra3 = new_ra3.wrap_at(180*u.degree)
new_dec3 = coord.Angle(df3["decg"],unit=u.degree)
```

```python
field1000_ra_deg=327.620330810547 
field1000_dec_deg=-38.6540641784668
field1000_ra=coord.Angle(field1000_ra_deg,unit=u.degree)
field1000_ra.degree
field1000_ra = field1000_ra.wrap_at(180*u.degree)
field1000_dec = coord.Angle(field1000_dec_deg,unit=u.degree)
```

```python
import astropy.coordinates as coord
import astropy.units as u
from astropy.coordinates import Angle, SkyCoord
from regions import PixCoord, CircleSkyRegion, CirclePixelRegion, RectangleSkyRegion, RectanglePixelRegion
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="mollweide")
#ax = plt.gca()
ax.scatter(new_ra1.radian, new_dec1.radian,color="r",label="SNLS D1")
ax.scatter(new_ra2.radian, new_dec2.radian,color="b",label="SNLS D2")
ax.scatter(new_ra3.radian, new_dec3.radian,color="g",marker='o',label="SNLS D4")
ax.scatter(field1000_ra.radian, field1000_dec.radian,color="k",label='Cadence Field1000')
ax.set_xticklabels(['14h','16h','18h','20h','22h','0h','2h','4h','6h','8h','10h'])
ax.grid(True)
ax.legend()
plt.show()
```
