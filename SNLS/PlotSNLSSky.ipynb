```python
new_ra = coord.Angle(df["rag"], unit=u.degree) # create an Angle object
new_ra.degree # convert to degrees
new_ra = new_ra.wrap_at(180*u.degree)
new_dec = coord.Angle(df["decg"]*u.degree)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="mollweide")
#ax = plt.gca()
ax.scatter(new_ra.radian, new_dec.radian,color="r")
ax.set_xticklabels(['14h','16h','18h','20h','22h','0h','2h','4h','6h','8h','10h'])
ax.grid(True)
plt.show()
```
