```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
files=['D1_ab_v3.1.list','D2_ab_v3.1.list','D4_ab_v3.1.list']
```

```python
colname =[
"rag","decg",\
    "mxu","mx0u","emagu","nmeasu","mean_xfpu","mean_yfpu","chi2u",\
    "mxg","mx0g","emagg","nmeasg","mean_xfpg","mean_yfpg","chi2g",\
    "mxr","mx0r","emagr","nmeasr","mean_xfpr","mean_yfpr","chi2r",\
    "mxi","mx0i","emagi","nmeasi","mean_xfpi","mean_yfpi","chi2i",\
    "mxi2","mx0i2","emagi2","nmeasi2","mean_xfpi2","mean_yfpi2","chi2i2",\
    "mxz","mx0z","emagz","nmeasz","mean_xfpz","mean_yfpz","chi2z"] 
```

```python
colname_keep =[
"rag","decg",\
    "mxu",\
    "mxg",\
    "mxr",\
    "mxi",\
    "mxi2",\
    "mxz"] 
```

# Read files

## Read Dataset 1

```python
df1=pd.read_table(files[0],sep=' ' ,skiprows=67,names=colname,index_col=False)
```

```python
#df1.head()
```

```python
df1keep=df1.loc[:,colname_keep]
```

```python
df1keep["data"]='D1'
```

```python
df1keep.head()
```

## Read Dataset 2

```python
df2=pd.read_table(files[1],sep=' ' ,skiprows=67, names=colname,index_col=False)
```

```python
df2keep=df2.loc[:,colname_keep]
```

```python
df2keep["data"]='D2'
```

```python
df2keep.head()
```

## Read Dataset 4

```python
df3=pd.read_table(files[2],sep=' ' ,skiprows=67,names=colname,index_col=False )
```

```python
df3keep=df3.loc[:,colname_keep]
```

```python
df3keep["data"]='D4'
```

```python
df3keep.head()
```

# Merge the 3 datasets

```python
frames = [df1keep, df2keep, df3keep]
```

```python
df = pd.concat(frames)
```

```python
df
```

```python
df.dropna(axis=0, how='any',inplace=True)
```

```python
df
```

```python
df.plot.scatter(x='mxu', y='mxg',grid=True,figsize=(8,8),color='r');
```

```python
arr_ug=df["mxu"]-df["mxg"]
arr_gr=df["mxg"]-df["mxr"]
```

```python
arr=np.column_stack((arr_ug, arr_gr))
```

```python
arr.shape
```

```python
df_col=pd.DataFrame(data=arr,
...                    columns=['u-g', 'g-r'])
```

```python
df_col.head()
```

```python
df_col.plot.scatter(x="u-g", y="g-r",grid=True,figsize=(8,8),color='b',xlim=(-2,4));
```
