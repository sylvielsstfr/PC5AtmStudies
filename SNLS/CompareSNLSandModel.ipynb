```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# to enlarge the sizes
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (10, 10),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
plt.rcParams.update(params)
```

# input file

```python
file_SNLS='SNLS_col_D1D2D4.csv'
```

```python
#file_phoenix='phoenix_snlsmag_new.csv'
#file_phoenix='phoenix_snlsmag_extinct_50.csv'
file_phoenix='phoenix_snlscol_Noextinct_2.csv'
```

# Read file

```python
snls=pd.read_csv(file_SNLS,index_col=0)
phoenix=pd.read_csv(file_phoenix,index_col=0)
```

```python
snls.head()
```

```python
phoenix.head()
```

```python
plt.scatter(snls["u-g"],snls["g-r"],marker='.',color="red",label='snls')
plt.scatter(phoenix["u-g"],phoenix["g-r"],marker='.',color="blue",label='phoenix')
plt.grid()
plt.legend()
plt.xlabel('u-g',fontsize=16,fontweight='bold')
plt.ylabel('g-r',fontsize=16,fontweight='bold')
plt.title('2D color plot SNLS/Phoenix')
```

```python
plt.scatter(snls["g-r"],snls["r-i"],marker='.',color="red",label='snls')
plt.scatter(phoenix["g-r"],phoenix["r-i"],marker='.',color="blue",label='phoenix')
plt.grid()
plt.legend()
plt.xlabel('g-r',fontsize=16,fontweight='bold')
plt.ylabel('r-i',fontsize=16,fontweight='bold')
plt.title('2D color plot SNLS/Phoenix')
```

```python
plt.scatter(snls["r-i"],snls["i-z"],marker='.',color="red",label='snls')
plt.scatter(phoenix["r-i"],phoenix["i-z"],marker='.',color="blue",label='phoenix')
plt.grid()
plt.legend()
plt.xlabel('r-i',fontsize=16,fontweight='bold')
plt.ylabel('i-z',fontsize=16,fontweight='bold')
plt.title('2D color plot SNLS/Phoenix')
```

```python
fig, axes = plt.subplots(1, 3,figsize=(20,6))

axes[0].scatter(snls["u-g"],snls["g-r"],marker='.',color="red",label='snls')
axes[0].scatter(phoenix["u-g"],phoenix["g-r"],marker='.',color="blue",label='phoenix')
axes[0].grid()
axes[0].legend()
axes[0].set_xlabel('u-g',fontsize=16,fontweight='bold')
axes[0].set_ylabel('g-r',fontsize=16,fontweight='bold')

axes[1].scatter(snls["g-r"],snls["r-i"],marker='.',color="red",label='snls')
axes[1].scatter(phoenix["g-r"],phoenix["r-i"],marker='.',color="blue",label='phoenix')
axes[1].grid()
axes[1].legend()
axes[1].set_xlabel('g-r',fontsize=16,fontweight='bold')
axes[1].set_ylabel('r-i',fontsize=16,fontweight='bold')


axes[2].scatter(snls["r-i"],snls["i-z"],marker='.',color="red",label='snls')
axes[2].scatter(phoenix["r-i"],phoenix["i-z"],marker='.',color="blue",label='phoenix')
axes[2].grid()
axes[2].legend()
axes[2].set_xlabel('r-i',fontsize=16,fontweight='bold')
axes[2].set_ylabel('i-z',fontsize=16,fontweight='bold')
plt.suptitle('2D color plot SNLS/Phoenix, Extinct=0',fontsize=25,fontweight='bold')



plt.scatter(snls["r-i"],snls["i-z"],marker='*',color="red",label='snls')
plt.scatter(phoenix["r-i"],phoenix["i-z"],marker='*',color="blue",label='phoenix')
plt.grid()
plt.legend()
plt.xlabel('r-i',fontsize=16,fontweight='bold')
plt.ylabel('i-z',fontsize=16,fontweight='bold')
plt.title('2D color plot SNLS/Phoenix')
```
