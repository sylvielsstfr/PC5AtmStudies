```python
top_pysynphot_data_dir=os.environ['PYSYN_CDBS']
import pysynphot as S
```

```python
bp_u = S.ObsBandpass('sdss,u')
bp_g = S.ObsBandpass('sdss,g')
bp_r = S.ObsBandpass('sdss,r')
bp_i = S.ObsBandpass('sdss,i')
bp_z = S.ObsBandpass('sdss,z')
```

```python
plt.plot(bp_u.wave,bp_u.throughput,'b-')
plt.plot(bp_g.wave,bp_g.throughput,'g-')
plt.plot(bp_r.wave,bp_r.throughput,'r-')
plt.plot(bp_i.wave,bp_i.throughput,'-',color='orange')
plt.plot(bp_z.wave,bp_z.throughput,'k-')
plt.grid()
```
